```python
import asterix
import pyModeS as pms
import time
import pandas as pd
from pandas import ExcelWriter
import xlrd 

#Importamos el archivo de datos:
sample_filename = "./gdrive/My Drive/TFG/ArchivosAsterix/Test_Decoder2.xlsx"

wb = xlrd.open_workbook(sample_filename)
hoja = wb.sheet_by_index(0)

nrows = hoja.nrows
ncols = hoja.ncols

# for i in range(0,ncols):
#   if i != 1 and i != 3:
#     print(hoja.cell_value(1,i))


for i in range(1,2):
    print(f'Mensaje nº{i}:'.center(30,'-'))
    hexD = []
    dat = []
    lon = int(hoja.cell(i,3).value) + 2
    for j in range(lon): 
      fspec = pms.bin2hex(hoja.cell(i,4).value)
      if j > 4:
        hexD.append(hoja.cell(i,j).value)
      if j != 1 and j != 3 and j != 4:
        dat.append(hoja.cell(i,j).value)
      hexData = "".join(hexD)
      msg = "".join(dat)
    dato = bytearray.fromhex(msg)
    packed=asterix.parse(dato)
    print(f'Packet: {packed}')
    print(f'Fspec: {fspec}')
    print(f'HexData: {hexData}')
    print()

```

```python
from pytictoc import TicToc
import asterix
import time


sample_filename = "./gdrive/My Drive/TFG/ArchivosAsterix/ADSB_HEX_08.txt"
packed=[]

#Abrimos el archivo de datos y lo leemos:

with open(sample_filename, "rt") as f:

  data = f.readlines()
  n_msg=len(data)

  for k in range(1,500): #range(1,n_msg):
      dato = bytearray.fromhex(data[k])
      try: # Para evitar mensajes que hace que ASTERIX "explote"
          # Entrenar
          t = TicToc()
          t.tic()
          while t.tocvalue() < 0.001: 
            packed=asterix.parse(dato)
      except BaseException:
        print('Mensaje ', k, 'no decodificado ')

      if k % 100 == 0:
        print(round(k/n_msg*100,2),"% | ",k)

      time.sleep(0.05)

#print(packed)
```
